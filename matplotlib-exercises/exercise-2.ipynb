## Most Common 15 Languages

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter
```

```python
data = pd.read_csv("C:/Users/MSI/Desktop/data-science-projects/project-2/survey_results_public.csv")
```

```python
clean_lang_data = data[data["LanguageWorkedWith"].notna()]

# In "worked_languages", respondents' answers are combined. 
# Therefore, it includes duplicates of languages.
worked_languages = []

for respondent_languages in clean_lang_data["LanguageWorkedWith"]:
    respondent_languages = respondent_languages.split(";")
    for every_language in respondent_languages:
        worked_languages.append(every_language)
```

```python
lang_and_count = Counter(worked_languages).most_common()

languages = []
popularity = []

for every_tuple in lang_and_count:

    languages.append(every_tuple[0])
    popularity.append(every_tuple[1])
```

```python
# to make bar chart descending order
languages.reverse()
popularity.reverse()
```

```python
plt.style.use("fivethirtyeight")
plt.rcParams.update({'font.size': 6.5})

plt.barh(languages, popularity)
plt.title("Most Common 15 Languages")
plt.ylabel("Number of People Who Use")
plt.show()
```
